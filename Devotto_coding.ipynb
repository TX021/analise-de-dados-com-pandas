```python
#importando pacotes 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#lendo arquivo
df = pd.read_excel('./Lista_Devotto.xlsx')

```

```python
# 1.Gráfico de barras mostrando a distribuição das queixas por gênero
contagem_queixas = df.groupby('GÊNERO')['QUEIXA'].count()

# 2. Mostrando a distribuição de idade
distribuicao_idade = df['IDADE']

# 3. Gráfico de barras exibindo as reclamações mais comuns
frequencia_queixas = df['QUEIXA'].value_counts()

# Plot 1:Distribuição de queixas por gênero
plt.figure(figsize=(10, 10))
contagem_queixas.plot(kind='bar')
plt.title('Distribuição de Queixas por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Quantidade de Queixas')
plt.show()

# Plot 2: Distribuição de idade
plt.figure(figsize=(10, 10))
plt.hist(distribuicao_idade, bins=10, edgecolor='black')
plt.title('Distribuição por idade')
plt.xlabel('idade')
plt.ylabel('Frequência')
plt.show()

# Plot 3: Mostrando a frequência das queixas
plt.figure(figsize=(10, 10))
frequencia_queixas.plot(kind='bar')
plt.title('Frequência das Queixas')
plt.xlabel('Queixas')
plt.ylabel('Frequência')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
