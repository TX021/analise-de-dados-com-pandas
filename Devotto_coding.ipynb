```python
#importando pacotes 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#lendo arquivo
df = pd.read_excel('./Lista_Devotto.xlsx')

```

```python
# 1.Grfico de barras mostrando a distribuio das queixas por gnero
contagem_queixas = df.groupby('GNERO')['QUEIXA'].count()

# 2. Mostrando a distribuio de idade
distribuicao_idade = df['IDADE']

# 3. Grfico de barras exibindo as reclamaes mais comuns
frequencia_queixas = df['QUEIXA'].value_counts()

# Plot 1:Distribuio de queixas por gnero
plt.figure(figsize=(10, 10))
contagem_queixas.plot(kind='bar')
plt.title('Distribuio de Queixas por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Quantidade de Queixas')
plt.show()

# Plot 2: Distribuio de idade
plt.figure(figsize=(10, 10))
plt.hist(distribuicao_idade, bins=10, edgecolor='black')
plt.title('Distribuio por idade')
plt.xlabel('idade')
plt.ylabel('Frequncia')
plt.show()

# Plot 3: Mostrando a frequncia das queixas
plt.figure(figsize=(10, 10))
frequencia_queixas.plot(kind='bar')
plt.title('Frequncia das Queixas')
plt.xlabel('Queixas')
plt.ylabel('Frequncia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
